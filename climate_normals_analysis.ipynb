```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the data from the CSV file with tab delimiter
df_climate = pd.read_csv('Datasets/climate_normals_bbsr.csv', delimiter='\t')

df_climate.head()
```

```python
import calendar

# Create a dictionary to map month numbers to month names
month_map = {i: calendar.month_name[i] for i in range(1, 13)}

# Replace the month numbers with month names in the dataframe
df_climate['month'] = df_climate['month'].map(month_map)

df_climate.head()
```

```python
# Check for null values
print(df_climate.isnull().sum())

# Drop rows with any null values
df_climate = df_climate.dropna()
```

```python
# Find unique instances of weather types or show "Not available" if not exist
unique_tmin = df_climate['tmin'].dropna().unique() if not df_climate['tmin'].dropna().empty else "Not available"
unique_tmax = df_climate['tmax'].dropna().unique() if not df_climate['tmax'].dropna().empty else "Not available"
unique_prcp = df_climate['prcp'].dropna().unique() if not df_climate['prcp'].dropna().empty else "Not available"
unique_wspd = df_climate['wspd'].dropna().unique() if not df_climate['wspd'].dropna().empty else "Not available"
unique_pres = df_climate['pres'].dropna().unique() if not df_climate['pres'].dropna().empty else "Not available"
unique_tsun = df_climate['tsun'].dropna().unique() if not df_climate['tsun'].dropna().empty else "Not available"

print("Unique Tmin values:", unique_tmin)
print("Unique Tmax values:", unique_tmax)
print("Unique Prcp values:", unique_prcp)
print("Unique Wspd values:", unique_wspd)
print("Unique Pres values:", unique_pres)
print("Unique Tsun values:", unique_tsun)
```

```python
# Find the maximum and minimum temperatures over the years
max_temp = df_climate['tmax'].max()
min_temp = df_climate['tmin'].min()

# Print the results
print(f"The maximum temperature over the years is: {max_temp}")
print(f"The minimum temperature over the years is: {min_temp}")

# Analyze the temperatures
if pd.notna(max_temp) and pd.notna(min_temp):
    temp_range = max_temp - min_temp
    print(f"The temperature range over the years is: {temp_range}")
else:
    print("Temperature data is not available for analysis.")
```
